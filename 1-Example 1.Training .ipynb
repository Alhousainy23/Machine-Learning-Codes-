```python
#Example Calculate The Value Of Pi 
import math 
pi=0
for k in range(20):
    pi+=pow(-3,-k) / (2 * k + 1)
pi*=math.sqrt(12)
print('The Value Of Pi Is ==> ',pi)
print('Error Is ==> ',abs(pi - math.pi))
#==========================================================
print('='*40)
#Example Two To Calculate Pi Value 
v=0
for n in range(100):
    v+=pow(-3 ,-n) / (2 * n +1)
v*=math.sqrt(12)
print('The Value Of Pi Is ==> %.6f' %v)
print('The Error Is ==> ',abs(v-math.pi))
print('='*70)
#=======================================================================
#Calculate Value Pi Using Mont Carloo
import random as rn 
import numpy as np
x=[]
y=[]
def calcpi(n):
    counter =0
    for k in range(1,n+2):
        x.append(rn.random())
        y.append(rn.random())
    for gg in range(1,n+1):
        for jj in range(1,n+1):
            if ((x[gg])**2) +((y[jj])**2)<=1:
                counter+=1
    pii=(float(counter))/(n**2)
    return pii
calculatepi=4*calcpi(3000)
print('Calculate Pi Is = '+str(calculatepi))
print('Difference Between = '+str(np.pi -calculatepi))
```

```python
#Calculate X Value And Draw Sin Function 
import math as m 
import pylab as p 
xmin,xmax= -2.*m.pi , 2.*m.pi
n=1000
x=[0.]*n
y=[0.]*n
dx=(xmax-xmin)/(n-1)
for i in range(n):
    xpt=xmin+i*dx
    x[i]=xpt
    y[i]=m.pow(m.sin(xpt),2)
p.plot(x,y)
p.show()
```

```python
#Example To Calculate Divorce_rate 
import pylab as p
#Data Needed To Represnts And Draw  
years=range(2000,2010)
divorce_rate=[5.0,4.7,4.6,4.4,4.3,4.1,4.2,4.2,4.2,4.1]
margraine_consumption=[8.2,7,6.5,5.3,5.2,4,4.6,4.5,4.2,3.7]
#=================================================================
#Draw Line One (Years & divorce_rate)
line1=p.plot(years,divorce_rate,'b-o',label='Divorce Rate In Maine')
p.ylabel('Divorces Per 1000 People')
p.xlabel('Years')
p.legend()
#================================================================
p.twinx()#Used To Draw Two Graphs Together Contain (Same X Axsis But Different Y Axsis)
#==================================================================
#Draw Line Two (Years & Margrine Consumption)
line2=p.plot(years,margraine_consumption,'r-o',label='Margrine Cons')
p.ylabel('1b of Margraine(per Capita)')
#======================================================================
#Two Create Legend 
lines= line1 + line2
labels=[]
for line in lines:
    labels.append(line.get_label())
p.legend(lines,labels)
p.show()
```

```python
#Example To Calculate The Density Of Planets 
import math as m 
#Create Dictionar That Conatins Planeys Names and Values 
body={'Sun':(1.988e30,6.955e5),'Mercury':(3.301e23,2440.),'Venus':(4.867e+24,6052.),'Earth':(5.972e24,6371.),
'Mars':(6.417e23,3390.),'Jupiter':(1.899e27,69911.),'Saturn':(5.685e26,58232.),'Uranus':(8.682e25,25362.),'Neptune':(1.024e26,24622.)}
#Take Body Dictionary Keys And Put In Planets List
planets=list(body.keys())
planets.remove('Sun')#Remove Sun From List
#====================================================================
#Create Function To Calculate The Density For Planets
def calc_density(m1,r): return m1 / (4/3 * m.pi * m.pow(r,3))
#====================================================================
rho={}#Create Empty Dictionary 
#Create For Loop Becuase Take Value Of M and R to Send Function 
for planet in planets:
    m1,r=body[planet]
    #Calculate Density in g/cm3
    rho[planet]=calc_density(m1 * 1000 , r * 1.e5)
#====================================================================
#Create For Loop To Display The Density 
for planet , density in sorted (rho.items()): print('The Density of {0} Is ==> {1:3.2f} G/Cm3 '.format(planet,density))
```

```python
#Make Transpose Matrix 
#===============================================================================
#Make Function To Make Transpose Matrix
def makeT(mm):
    rr=len(mm)#To Know Number Of Rows
    cc=len(mm[0])#To Know Number Of Columns 
    tm=[[0 for i in range(rr)]for j in range(cc)]#To Make Matrix Contain 4 Rows And 5 Columns Values Is = Zero
    for cl in range(cc):
        for ro in range(rr):
            tm[cl][ro]=mm[ro][cl]
    return tm
#==================================================================================
m=[[1,20,3,4],[4,5,61,8],[7,8,9,60],[30,3,6,3],[0,7,60,9]]
print(makeT(m))
```
