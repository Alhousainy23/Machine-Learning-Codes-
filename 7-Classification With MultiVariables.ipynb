```python
#Make Sigmoid Function 
def sigmoid_func(num) : return 1 / (1 + np.exp(-num))
#==============================================================
#Make Cost Function 
def cost_function (theta , x , y , alpha):
    theta_values =np.matrix(theta)
    x_values = np.matrix(x)
    y_values = np.matrix(y)
    first_eq = np.multiply(- y_values , np.log(sigmoid_func(x_values * theta_values.T)))
    second_eq = np.multiply((1 - y) ,np.log((1- sigmoid_func(x_values * theta_values.T))))
    reg_eq = (alpha / 2 * len(x_values)) * np.sum(np.power(theta_values[:,1:theta_values.shape[1]] , 2))
    return np.sum (first_eq - second_eq) / len (x_values) +reg_eq
```

```python
def gradient_descent_with_loop(theta , x , y , alpha):
    theta_value=np.matrix(theta)
    x_value = np.matrix(x)
    y_value = np.matrix(y)
    parameters = int(theta_value.ravel().shape[1])
    grad=np.zeros(parameters)
    error =sigmoid_func(x_value * theta_value.T) - y_value
    for i in range (parameters):
        term=np.multiply(error ,x_value[: , i])
        if (i == 0): grad[i] = np.sum(term) / len (x_value)
        else : grad[i] = (np.sum(term) / len(x_value)) + (( alpha / len(x_value)) * theta_value[:,i] )
    return grad
```

```python
def GD(theta , x, y , lr):
    theta_values = np.matrix(theta)
    x_values = np.matrix(x)
    y_values = np.matrix(y)
    paramters = int (theta_values.ravel().shape[1])
    error = sigmoid_func(x_values * theta_values.T) -y_values
    grad = ((x_values.T * error) / len(x_values)).T + (( lr / len(x_values)) * theta_values)
    #Intercept gradient is not reguarlized
    grad [0,0] = np.sum(np.multiply(error , x_values[:,0])) / len(x_values)
    return np.array(grad).ravel()
```

```python
from scipy.optimize import minimize
def one_vs_all (x , y , num_labels , learning_rate):
    rows =x.shape[0]#5000
    params =x.shape[1]#400
    #K X (N +1) array for paramters of each of the k classifiers
    all_theta =np.zeros((num_labels,params+1))
    print('All Theta Shape ==> ',all_theta.shape)#
    #Insert a column of ones at the beginning for the intercept term 
    x = np.insert(x,0,values=np.ones(rows),axis =1)
    print('The Shape Of Matrix X Is ==> ',x.shape)#
    #Labels Are 1 -Indeeded Instead of 0-Indeeded
    for i in range( 1 , num_labels + 1):
        theta =np.zeros(params + 1)
        y_i = np.array( [ 1 if label == i else  0 for label in y])
        y_i = np.reshape(y_i , (rows ,1))
        #Minimize the objective function 
        fmin = minimize(fun=cost_function , x0 =theta ,args=(x ,y_i , learning_rate),method='TNC',jac =GD)
        all_theta[i-1:] =fmin.x
    return all_theta
```

```python
all_theta = one_vs_all(data['X'],data['y'] , 10 , 1)
print('The All Theta Shape Is ==> ',all_theta.shape)#
print('\t\t The All Theta Is \n',all_theta)
def prdict_all (x, all_theta):
    rows = x.shape[0]
    params =x.shape[1]
    num_labels = all_theta.shape[0]
    #Same as before , insert ones to match the shape 
    x = np.insert(x , 0 , values= np.ones(rows),axis =1)
    #Convert To Matrix Because To Easily Treatment
    x_new = np.matrix(x)
    all_theta_new = np.matrix(all_theta)
    #Compute The Class Probability For Each Class on Each Training Instance 
    h =sigmoid_func(x_new * all_theta_new.T)
    #Create Array of the index with the maximum probability 
    h_argmax = np.argmax(h , axis = 1)
    #Because Our Array was zero-indexed we need to add one for the true label prediction 
    h_argmax =h_argmax + 1
    return h_argmax
```

```python
y_pred = prdict_all(data['X'],all_theta)
correct = [ 1 if a == b else 0 for (a , b ) in zip (y_pred , data ['y'])]
accuracy = (sum (map ( float ,correct)) / float (len (correct)))
print('The Accuracy Is = {0} % '.format(accuracy * 100))
```
